### 3. Extracting SRA Files to RAW FASTQ format
Since our downstream tools (like `FastQC` and `Bowtie2`) cannot read the proprietary `.sra` format, we use NCBI's `fasterq-dump` to extract the raw sequencing reads. 

Given that the data comes from an Illumina NovaSeq 6000 platform, it is **paired-end**. We use a Bash loop to dynamically navigate through each directory, locate the `.sra` file, and extract them into forward (`_1.fastq`) and reverse (`_2.fastq`) strands.

* **Key Flags used:**
  * `--outdir raw_data/`: Saves the extracted files directly into our raw data directory.

````python
%%bash
for SRA_DIR in raw_data/SRR*; do
    # Extract the directory name (e.g., SRR32812560)
    BASE_NAME=$(basename $SRA_DIR)
    
    # Define the exact path to the .sra file
    SRA_FILE="$SRA_DIR/${BASE_NAME}.sra"   

    echo "Processing sample: ${BASE_NAME}..."
    
    # Run fasterq-dump with proper spacing
    fasterq-dump $SRA_FILE --outdir raw_data/ --split-files
done
echo "All samples extracted successfully!"
````
