# 📊 Notebook 02: Quality Control of Raw RNA-Seq Data

## 🧬 Introduction to Quality Control (QC)
Before aligning our sequencing reads to the *Mycobacterium tuberculosis* reference genome, it is crucial to perform quality control. Raw data from high-throughput sequencers often contains low-quality bases, sequencing artifacts, or adapter contamination that can bias downstream analysis.

In this notebook, we utilize **FastQC**, an industry-standard quality control tool, to inspect our raw `.fastq.gz` samples.

### Objectives:
1. Generate quality control reports for all 6 compressed samples.
2. Store the output HTML and ZIP files in a dedicated results directory.

```python
%%bash
fastqc raw_data/*.fastq.gz -o results/fastqc_results
```

### ⚙️ Step 2: Quality Trimming and Filtering using fastp
We use `fastp` to clean our raw data. This ultra-fast tool will automatically:
1. Trim low-quality bases from the ends where the quality drops.
2. Filter out reads that are too short after trimming.
3. Generate detailed HTML/JSON quality reports post-trimming.

All cleaned files will be stored in `results/trimmed_data/`.

```python
%%bash
# 1. Create a directory for the trimmed outputs
mkdir -p results/trimmed_data

# 2. Run the loop over all compressed files
for FILE_PATH in raw_data/*.fastq.gz; do
    
    # Extract only the sample name (e.g., SRR32812560)
    BASE_NAME=$(basename "$FILE_PATH" .fastq.gz)
    
    echo "⚡ Trimming sample: ${BASE_NAME}..."
    
    # Run fastp with input (-i) and output (-o) flags
    fastp \
      -i raw_data/${BASE_NAME}.fastq.gz \
      -o results/trimmed_data/${BASE_NAME}_trimmed.fastq.gz \
      -h results/trimmed_data/${BASE_NAME}.html \
      -j results/trimmed_data/${BASE_NAME}.json \
      --thread 4
      
done
echo "🎉 All samples have been successfully trimmed!"
```

## 📌 Conclusion & Next Steps
The Quality Control and Trimming phase is complete:
1. **Raw QC (FastQC):** Confirmed high base quality (Q30+) but revealed a high GC content (~60-65%), which naturally aligns with the biological characteristics of *Mycobacterium tuberculosis*. The double peak in GC content indicates potential ribosomal RNA (rRNA) carryover or high expression of stress-response genes due to Ethambutol treatment.
2. **Trimming (fastp):** Successfully removed the low-quality trailing bases at the ends of the reads. The post-trimming HTML reports confirm that the read quality is now pristine and stable across all positions.

The trimmed datasets are stored in `results/trimmed_data/` and are fully optimized for the next major milestone: **Genome Alignment (Mapping) using Bowtie2**.
