# 📊 Notebook 04: Gene Expression Quantification

## 🎯 Introduction
Following the successful alignment of our RNA-Seq reads to the *Mycobacterium tuberculosis* reference genome, we now move to the quantification phase. In this notebook, we will count how many reads map to each gene feature to build our raw expression matrix.

### Objectives for this Notebook:
1. Install the **Subread** package to access the `featureCounts` tool.
2. Run `featureCounts` using our sorted `.bam` files and the reference `.gff` annotation file.
3. Generate a clean raw counts table for downstream Differential Expression Analysis (DEA).

```python
%%bash
# 1. create folder
mkdir -p results/count_matrix

# 2. Run FeatureCounts 
featureCounts \
  -a reference/M_tuberculosis_H37Rv.gff \
  -F GFF \
  -t gene \
  -g locus_tag \
  -o results/count_matrix/raw_counts.txt \
  -T 4 \
  results/mapped_data/*_sorted.bam
```

## 📌 Conclusion
The Gene Quantification phase using `featureCounts` was completed with high efficiency:
* **Assignment Rate:** Successfully assigned ~**95%** of the aligned reads to specific *M. tuberculosis* locus tags (e.g., `Rv0001`, `Rv0002`).
* **Output Matrix:** Generated a comprehensive expression matrix (`raw_counts.txt`) containing raw integer counts across all 6 samples.
* **Data Quality:** The extremely high assignment rate confirms excellent annotation overlap and minimal structural alignment ambiguity.

The raw counts matrix is now ready as the primary input for downstream Differential Expression Analysis (DEA).
